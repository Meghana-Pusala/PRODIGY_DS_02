# Task2

Cleaning the data by handling the missing values and then proceed with exploratory data analysis (EDA)

```python
import pandas as pd

# Load the dataset
file_path = 'Task2.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Display a concise summary of the dataset
print("\nSummary of the dataset:")
print(df.info())

# Display descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
print(df.describe())

# Display descriptive statistics for categorical columns
print("\nDescriptive statistics for categorical columns:")
print(df.describe(include=['O']))  # 'O' refers to object data types (categorical variables)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming the DataFrame df is already loaded and cleaned

# Distribution of Age
plt.figure(figsize=(6, 4))
plt.hist(df['Age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Distribution of Fare
plt.figure(figsize=(6, 4))
plt.hist(df['Fare'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Count of Passengers by Class
plt.figure(figsize=(6, 4))
class_counts = df['Pclass'].value_counts().sort_index()
plt.bar(class_counts.index, class_counts.values, color='lightcoral', edgecolor='black')
plt.title('Count of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.xticks(class_counts.index)
plt.grid(True)
plt.show()
```

```python
# Count of Passengers by Gender
plt.figure(figsize=(6, 4))
gender_counts = df['Sex'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color='lightblue', edgecolor='black')
plt.title('Count of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Pairplot replacement - scatter plots and histograms
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# Scatter plots
axs[0, 1].scatter(df['Age'], df['Fare'], alpha=0.5, color='purple')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Fare')
axs[0, 1].set_title('Age vs Fare')

axs[1, 0].scatter(df['Age'], df['SibSp'], alpha=0.5, color='orange')
axs[1, 0].set_xlabel('Age')
axs[1, 0].set_ylabel('SibSp')
axs[1, 0].set_title('Age vs SibSp')

axs[1, 2].scatter(df['Fare'], df['SibSp'], alpha=0.5, color='green')
axs[1, 2].set_xlabel('Fare')
axs[1, 2].set_ylabel('SibSp')
axs[1, 2].set_title('Fare vs SibSp')

# Histograms
axs[0, 0].hist(df['Age'], bins=30, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Age Distribution')

axs[1, 1].hist(df['Fare'], bins=30, color='lightgreen', edgecolor='black')
axs[1, 1].set_title('Fare Distribution')

axs[2, 0].hist(df['SibSp'], bins=8, color='lightcoral', edgecolor='black')
axs[2, 0].set_title('SibSp Distribution')

axs[2, 2].hist(df['Parch'], bins=6, color='lightblue', edgecolor='black')
axs[2, 2].set_title('Parch Distribution')

# Remove unused subplots
axs[0, 2].remove()
axs[2, 1].remove()

plt.tight_layout()
plt.show()
```
